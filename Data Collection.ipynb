```python
import pandas as pd
```

```python
data = pd.read_csv("cpdata.csv")
fert = pd.read_csv("Fertilizer.csv")
```

```python
def low(x):
    return str(x).lower()
```

```python
data["label"] = data["label"].apply(low) # converting in to lower
fert["Crop"] = fert["Crop"].apply(low)
```

```python
la = data.label.unique()
```

```python
la
```

```python
mt= []
dic = []
for l in la:
    mt.append(fert[fert["Crop"]== l])
    if len(fert[fert["Crop"]== l]) > 0:
        dic.append(l)
```

```python
dic
```

```python
lt = []
for m in dic:
    lt.append(fert[fert["Crop"] == m])
    print(m)
```

```python
lt[14]
```

```python
final = pd.DataFrame()
```

```python
for i in range(0,15):
    c = pd.concat([final,lt[i]])
    final = c
```

```python
final_of_fert = final.groupby(by = "Crop")["N","P","K","pH"].mean()
```

```python

#final_of_fert.columns = ["label", "N" , "P","K","ph" ]
pd.DataFrame(final_of_fert)
```

```python
data.rename(columns={"label": "Crop"}, inplace=True)
```

```python
fuck = pd.merge(data,final_of_fert,on="Crop")
```
